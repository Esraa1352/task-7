```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'D:\MIA\task 4\Weather Dataset.csv')
df= df.dropna()
df= df.drop_duplicates()
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'],utc=True)
df = df.sort_values(by='Formatted Date',ascending=True)
df=df[df['Pressure (millibars)']>900]
df=df[df['Temperature (C)']>-90]
df=df[df['Humidity']>0]
df.to_csv(r'D:\MIA\task 4\Processed_Weather_Dataset.csv', index=False)
```

```python
plt.plot(df['Formatted Date'] , df['Temperature (C)'])
plt.xlabel('time')
plt.ylabel('Temperature')
plt.show()  
```
